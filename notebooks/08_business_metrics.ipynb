```python
from xgboost import XGBClassifier
import pandas as pd
import json
```

```python
train = pd.read_csv("../data/interim/train.csv")
test = pd.read_csv("../data/interim/test.csv")
train.head()
```

```python
X_train = train.drop(columns=['readmitted_30d'])
y_train = train[['readmitted_30d']]
X_test = test.drop(columns=['readmitted_30d'])
y_test = test[['readmitted_30d']]
```

```python
with open("../artifacts/best_params.json","r") as f:
    best_params = json.load(f)

model = XGBClassifier(**best_params)
model.fit(X_train, y_train)
y_pred_proba = model.predict_proba(X_test)[:, 1]
```

```python
y_pred = (y_pred_proba >= 0.6).astype(int)
y_pred
```

```python
def calculate_business_metrics(y_true, y_pred_proba, threshold=0.6):
    y_pred = (y_pred_proba >= threshold).astype(int)
    y_true = y_true.squeeze()
    
    tp = ((y_pred == 1) & (y_true == 1)).sum()  # Berhasil teridentifikasi
    fp = ((y_pred == 1) & (y_true == 0)).sum()  # Salah prediksi
    fn = ((y_pred == 0) & (y_true == 1)).sum()  # Salah prediksi
    tn = ((y_pred == 0) & (y_true == 0)).sum()  # Berhasil teridentifikasi
    
    # Business metrics
    avg_readmission_cost = 4_684_050 
    intervention_cost = 500_000
    intervention_success_rate = 0.25 
    
    prevented_readmissions = tp * intervention_success_rate
    cost_savings = prevented_readmissions * avg_readmission_cost
    intervention_costs = (tp + fp) * intervention_cost
    net_savings = cost_savings - intervention_costs
    roi = (net_savings / intervention_costs) * 100
    
    print(f"Prevented Readmissions: {prevented_readmissions:.0f}")
    print(f"Cost Savings: Rp {cost_savings:,.0f}")
    print(f"Net Savings: Rp {net_savings:,.0f}")
    print(f"ROI: {roi:.1f}%")
    print(f"Precision: {tp/(tp+fp):.4f}")
    print(f"Recall: {tp/(tp+fn):.4f}")
    
    return {
        'roi': roi,
        'net_savings': net_savings,
        'precision': tp/(tp+fp),
        'recall': tp/(tp+fn)
    }

metrics = calculate_business_metrics(y_test, y_pred_proba)
```
