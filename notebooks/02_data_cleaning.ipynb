```python
with pd.io.stata.StataReader('../data/stata/DM2021_fktpkapitasi.dta') as reader:
    val_labels = reader.value_labels()
    df_fktp_decoded = reader.read(convert_categoricals=False)
    lbllist = reader._lbllist

def fix_duplicate_labels(label_dict):
    seen = {}
    fixed = {}
    for code, label in label_dict.items():
        if label in seen:
            seen[label] += 1
            fixed[code] = f"{label}_{seen[label]}"
        else:
            seen[label] = 0
            fixed[code] = label
    return fixed

fixed_labels = {k: fix_duplicate_labels(v) for k, v in val_labels.items()}

for col_idx, label_name in enumerate(lbllist):
    if label_name and label_name in fixed_labels:
        col = df_fktp_decoded.columns[col_idx]
        df_fktp_decoded[col] = df_fktp_decoded[col].map(fixed_labels[label_name]).astype('category')

df_fktp_decoded.head()
```

```python
df_fktp_decoded.info()
```

```python
df_fktp_decoded.columns = new_column_name
df_fktp_decoded.head()
```

```python
df_fktp_decoded.info()
```

```python
df_fktp_decoded.isnull().sum()
```

```python
df_fktp_decoded['kode_nama_diagnosis_ICD10'] = df_fktp_decoded['kode_nama_diagnosis_ICD10'].cat.add_categories(['Missing'])
```

```python
df_fktp_decoded['kode_nama_diagnosis_ICD10'] = df_fktp_decoded['kode_nama_diagnosis_ICD10'].fillna('Missing')
df_fktp_decoded.info()
```

```python
df_fktp_decoded['kode_diagnosis_ICD10'] = df_fktp_decoded['kode_diagnosis_ICD10'].replace('', '999')
```

```python
df_fktp_decoded.to_csv("../data/cleaned/fktp-decoded.csv", index=False)
```
